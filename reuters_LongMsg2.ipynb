### Red in neg/pos dictionaries

```python
pos_words = pd.read_pickle('DicPositiveWords.pickle')
neg_words = pd.read_pickle('DicNegativeWords.pickle')
```

### Do bag of words

```python
import helpers
```

```python
date_file
```

```python
%%time
reu_file_list = [f for f in os.listdir(PATH+'Reuters_US/') if '_2011' in f]
media_attention = []

for date_file in reu_file_list[:]:
    print(date_file)
    df_tmp = None
    try:
        df_tmp = pd.read_csv(PATH+'Reuters_US/'+date_file,
                             delimiter=';', 
                             parse_dates=[0], 
                             names=['Time', 'Short_Msg', 'Long_Msg', 'Link']
                            ).iloc[:,[0,2]].dropna()
        if isinstance(df_tmp.set_index('Time').index, pd.DatetimeIndex):
            if df_tmp is not None:
                df_tmp = df_tmp.set_index('Time')
                df_tmp2 = helpers.count_sensitive_words(df_tmp.copy(deep=True), 
                                                    ["Morgan Stanley", "Goldman Sachs", "BlackRock"]+[{'positive': pos_words[:70].values.tolist()},
                                                                   {'negative': neg_words[:70].values.tolist()}
                                                                  ], 
                                                    companies_neg_pos=["Morgan Stanley", "Goldman Sachs", "BlackRock"], str_col='Long_Msg')
                media_attention.append(df_tmp2)
        else:
            print('{}: can not convert time into datetime'.format(date_file))
    except Exception as e:
        print('{}: retrying... ({})'.format(date_file, e))
        try:
            df_tmp = pd.read_csv(PATH+'Reuters_US/'+date_file,
                                 delimiter=';', 
                                 parse_dates=[0], 
                                 names=['Time', 'Short_Msg', 'Long_Msg', 'Link','N']
                                ).iloc[:,[0,2]].dropna()
            if isinstance(df_tmp.set_index('Time').index, pd.DatetimeIndex):
                    if df_tmp is not None:
                        df_tmp = df_tmp.set_index('Time')
                        df_tmp2 = helpers.count_sensitive_words(df_tmp.copy(deep=True), 
                                                            ["Morgan Stanley", "Goldman Sachs", "BlackRock"]+[{'positive': pos_words[:70].values.tolist()},
                                                                           {'negative': neg_words[:70].values.tolist()}
                                                                          ], 
                                                            companies_neg_pos=["Morgan Stanley", "Goldman Sachs", "BlackRock"], str_col='Long_Msg')
                        media_attention.append(df_tmp2)  
                
            else:
                print('{}: can not convert time into datetime'.format(date_file))
        except Exception as e:
            print('{}: do not know what to do... ({})'.format(date_file, e))
        
    
    #break
```
