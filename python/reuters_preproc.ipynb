```python
reu_file_list = [f for f in os.listdir(PATH+'Reuters_US/') if '_20' in f]

df_reu_list = []
for date_file in reu_file_list:
    try:
        df_tmp = pd.read_csv(PATH+'Reuters_US/'+date_file,
                             delimiter=';', 
                             parse_dates=[0], 
                             names=['Time', 'Short_Msg', 'Long_Msg', 'Link']
                            ).iloc[:,:2].dropna()
        if isinstance(df_tmp.set_index('Time').index, pd.DatetimeIndex):
            df_reu_list.append(df_tmp)
        else:
            print('{}: can not convert time into datetime'.format(date_file))
    except Exception as e:
        print('{}: retrying... ({})'.format(date_file, e))
        try:
            df_tmp = pd.read_csv(PATH+'Reuters_US/'+date_file,
                                 delimiter=';', 
                                 parse_dates=[0], 
                                 names=['Time', 'Short_Msg', 'Long_Msg', 'Link','N']
                                ).iloc[:,:2].dropna()
            if isinstance(df_tmp.set_index('Time').index, pd.DatetimeIndex):
                df_reu_list.append(df_tmp)
            else:
                print('{}: can not convert time into datetime'.format(date_file))
        except Exception as e:
            print('{}: do not know what to do... ({})'.format(date_file, e))
    
df_reu = pd.concat(df_reu_list)
df_reu.to_csv(PATH+'reuters_ALL.csv.gz', compression='gzip', index=False)
```
